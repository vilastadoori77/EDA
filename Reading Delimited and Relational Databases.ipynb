<h2> Getting Data From Relational Databases</h2>
<p>
Data is commonly stored in RDBMS, and it is easy to get it into Python. We will use the most common one - MySQL

There are many libraries to connect MySQL and Python, such as pymysql, MySQLdb, etc. 
All of them follow the following procedure to connect to MySQL:
</p>
    
<ul>
<li>
Create a connection object between MySQL and Python</li>
<li>Create a cursor object (you use the cursor to open and close the connection)</li>
<li>Execute the SQL query</li>
<li>Retrieve results of the query using methods such as fetchone(),  fetchall(), etc.</li> </ul>

```python
import chardet
import pandas as pd

def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc

encoding_used = find_encoding("D:/upgrad/adult.csv")
print("encoding_used", encoding_used)
df_file = pd.read_csv("D:/upgrad/adult.csv", sep=',',encoding=encoding_used)
df_file.head()
df_file.columns
#df_file.pivot_table(values="Age", index="Country", columns="salary")
```
