```python
s = 'cyqfjhcclkbxpbojgkar'
from itertools import count

def long_alphabet(input_string):
    maxsubstr = input_string[0:0] # empty slice (to accept subclasses of str)
    for start in range(len(input_string)): # O(n)
        for end in count(start + len(maxsubstr) + 1): # O(m)
            substr = input_string[start:end] # O(m)
        if len(substr) != (end - start): # found duplicates or EOS
            break
        if sorted(substr) == list(substr):
            maxsubstr=substr
    return maxsubstr

bla = (long_alphabet(s))
print("Longest substring in alphabetical order is: %s" %bl)
```

```python
s = 'cyqfjhcclkbxpbojgkar'
longest = ""
max =""

for i in range(len(s) -1):
    if(s[i] <= s[i+1] ):
       longest = longest + s[i]
       if(i==len(s) -2):
           longest = longest + s[i+1]
    else:
        longest  = longest + s[i]        
        if(len(longest) > len(max)):
            max = longest
        longest = ""        

if(len(s) == 1):
    longest = s


if(len(longest) > len(max)):
    print("Longest substring in alphabetical order is: " + longest)
else:
    print("Longest substring in alphabetical order is: " + max)
```

```python
def median(a, l, r): 
    n = r - l + 1
    n = (n + 1) // 2 - 1
    return n + l 


def IQR(a, n): 
    a.sort() 
  
    # Index of median of entire data 
    mid_index = median(a, 0, n) 
    print("Q2", mid_index)
  
    # Median of first half 
    Q1 = a[median(a, 0, mid_index)] 
    print("Q1",Q1)
    # Median of second half 
    Q3 = a[median(a, mid_index + 1, n)] 
    print("Q3",Q3)
    # IQR calculation 
    return (Q3 - Q1)

list_shares = list(df[' shares'])
leng = len(list_shares)
iqr = IQR(list_shares,leng)
print(iqr)
```

```python
import pandas as pd 
pd.set_option('display.max_columns', 500)
import sys,ast
input_str = ['Serial No.', 'GRE Score', 'TOEFL Score']

def normalize(df):
    result = df.copy()
    for input_str in df.columns:
        max_value = df[input_str].max()
        min_value = df[input_str].min()
        result[input_str] = (df[input_str] - min_value) / (max_value - min_value)
    return result

df=pd.read_csv("https://media-doselect.s3.amazonaws.com/generic/GVj4YrNpJwOvL7gBe2B53wJy/Admission_Predict.csv")

df_normalize = normalize(df)
print(df_normalize.describe())
```
