```python
#my list
my_list=["apples", "cheery", "oranges", "Kiwi"]
print(my_list)

for fruit in my_list:
    if(fruit=="apples"):
        print("Apple a day keeps the doctor away")
    elif(fruit=="cheery"):
        print("Cherries are red in color")
    elif(fruit=="oranges"):
        print("THey are orange in color")
    else:
        print("wrong entry: Not sure about this fruit")
        
```
