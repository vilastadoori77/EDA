```python
import pandas as pd
import glob
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',500)
path = r'D:\upgrad\statsandEDA\gdp\NAD' # use your path
all_files = glob.glob(path + "/*.csv")
li=[]
for i, f in enumerate (all_files):
    if i == 0:        
        df = pd.read_csv(f, index_col=None, header=0, encoding = 'unicode_escape')
        fname = str(all_files[i]).split('\\')[-1:][0]
        fname1 = str(fname).split('-')[1]
        #print(fname1)
        df['state_name']=fname1
        df2= df[['Item','state_name','2014-15']]
        li.append(df2)
        
    else:
        tmp = pd.read_csv(f, index_col=None, header=0, encoding = 'unicode_escape')
        fname = str(all_files[i]).split('\\')[-1:][0]
        fname1 = str(fname).split('-')[1]
        #print(fname1)
        tmp['state_name']=fname1
        df3= tmp[['Item','state_name','2014-15']]
        li.append(df3)
        
frame = pd.concat(li, axis=0, ignore_index=True, sort=True)
frame
```
