```python
import pandas as pd 
pd.set_option('display.max_columns', 500)

url='https://media-doselect.s3.amazonaws.com/generic/eJ5zayg0OnnX5PqMvAR8KQd9p/titanicmod.csv'

df=pd.read_csv(url)

df
```

```python
df_remove=df[df.isnull().sum(axis=1) < 3]
df_remove
```

```python
nunique = df_remove.apply(pd.Series.nunique)
cols_to_drop = nunique[nunique == 1].index
df_remove=df_remove.drop(cols_to_drop, axis=1)
df_remove
```

```python
count=0
checkcount=0
for index, row in df_remove.iterrows():
    if(row.Year<1900):
        count = count+1
    if(row.Year>2018):
        checkcount = checkcount+1

print(checkcount+count)
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
cols = list(df_remove.select_dtypes(include=numerics).columns)
cols
df_normalize = df_remove[cols].apply(lambda x:(x-x.min()) / (x.max()-x.min()))
print(df_normalize.head(4))
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
cols = list(df_remove.select_dtypes(include=numerics).columns)
print(cols)
for column in df_remove:
    if column in cols:
        df_normalized=df_remove[cols].apply(lambda x:(x-x.min()) / (x.max()-x.min()))
        
df_normalized

```

```python
def normalize(df):
    result = df.copy()
    dtypes = list(zip(df.dtypes.index, map(str, df.dtypes)))
    for feature_name, dtype in dtypes:
        if((dtype=='int64')|(dtype=='float64')):
            max_value = df[feature_name].max()
            min_value = df[feature_name].min()
            result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

df_std_1 = normalize(df_remove)
print(df_std_1.head(3))
```
