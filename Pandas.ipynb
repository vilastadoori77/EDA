```python
# Creating some bunch of data frames as below:
import numpy as np
import pandas as pd

market_df = pd.read_csv("D:/upgrad/datasets/market_fact.csv")
customer_df = pd.read_csv("D:/upgrad/datasets/cust_dimen.csv")
product_df =pd.read_csv("D:/upgrad/datasets/prod_dimen.csv")
shipping_df=pd.read_csv("D:/upgrad/datasets/shipping_dimen.csv")
orders_df=pd.read_csv("D:/upgrad/datasets/orders_dimen.csv")

```

```python
#Merging the data
#how="inner"
#on="cust_id"

merge_cust_ord_df = pd.merge(market_df, customer_df, how="inner", on="Cust_id")
merge_cust_ord_df.head(100)
```

```python
# orders made by customers from corporate

merge_cust_ord_df.loc[merge_cust_ord_df['Customer_Segment']=='CORPORATE', :]
```

```python
# Order from product Category = office supplies and from corporate segment.
func_1 = pd.merge(merge_cust_ord_df, product_df, how="inner", on="Prod_id" )
func_1.head(200)

func_1.loc[(func_1['Product_Category']=='OFFICE SUPPLIES') & (func_1['Customer_Segment'] == 'CORPORATE'), :]
```

```python
# Merge Shipping Data

func1_ship = pd.merge(func_1,shipping_df,  how="inner", on="Ship_id")
func1_ship.head()

func_ship_custid_prdcat=func1_ship.loc[(func1_ship['Customer_Segment']=='CORPORATE')& (func1_ship['Product_Category']=='OFFICE SUPPLIES'),:]

```

```python

#merging the order data to create a master data
func_ship_custid_prdcat_ord = pd.merge(func_ship_custid_prdcat,orders_df, how='inner', on='Ord_id')
func_ship_custid_prdcat_ord
```

```python
# Teamwise stats for IPL 2018

IPL_2018 = pd.DataFrame({'IPL Team': ['CSK', 'SRH', 'KKR', 'RR', 'MI', 'RCB', 'KXIP', 'DD'],
                         'Matches Played': [16, 17, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [11, 10, 9, 7, 6, 6, 6, 5]}
                       )

IPL_2018.set_index('IPL Team', inplace=True)
IPL_2018
```

```python
IPL_2017 = pd.DataFrame({'IPL Team': ['MI', 'RPS', 'KKR', 'SRH', 'KXIP', 'DD', 'GL', 'RCB'],
                         'Matches Played': [17, 16, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [12, 10, 9, 8, 7, 6, 4, 3]})

IPL_2017.set_index('IPL Team', inplace=True)
IPL_2017                      
```

```python
Total = IPL_2018.add(IPL_2017, fill_value=0)
Total
```

```python
### Win percentage

Total['Win %'] = Total['Matches Won']/Total['Matches Played'] * 100
Total
```

```python
# Sorting based on asc is false

Total.sort_values(by=(['Matches Played', 'Matches Won']), ascending=False)
```

```python
#Given three data frames containing the number of gold, silver, and bronze Olympic medals won by some countries, 
#determine the total number of medals won by each country. 
#Note: All the three data frames dont have all the same countries. 
#So, ensure you use the fill_value argument (set it to zero), to avoid getting NaN values. 
#Also, ensure you sort the final dataframe, according to the total medal count in descending order.

gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )

gold.set_index('Country', inplace=True)
silver.set_index('Country', inplace=True)
bronze.set_index('Country', inplace=True)

Total_gold_silver = gold.add(silver, fill_value=0)
Total = Total_gold_silver.add(bronze, fill_value=0)
Total.sort_values(by=['Medals'], ascending=False)
```
